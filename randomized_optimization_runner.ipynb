```python
import mlrose_hiive as mlrose
import numpy as np
import pandas as pd
from functools import reduce

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.metrics import classification_report
from sklearn.metrics import log_loss


import seaborn as sns
```

```python
SEED = 7
OUTPUT_DIRECTORY = 'output'
```

# Four Peaks Fitness Function

```python
fp_fitness = mlrose.FourPeaks(t_pct=0.15)
fp_problem = mlrose.DiscreteOpt(length = 20, fitness_fn = fp_fitness, maximize = True, max_val = 2)
```

## Randomized Hill Climbing

```python
fp_rhc = mlrose.runners.RHCRunner(problem=fp_problem, experiment_name="fp_rhc", seed=SEED, 
                         iteration_list=[50, 100, 150], 
                         max_attempts=20, restart_list=[5, 25, 50, 100]) 

fp_rhc_run_stats, fp_rhc_run_curves = fp_rhc.run()
```

```python
sns.set(style='darkgrid')

sns.lineplot(x='Iteration', y='Fitness', data=fp_rhc_run_stats)
```

## Simulated Annealing

```python
fp_sa = mlrose.runners.SARunner(problem=fp_problem, experiment_name="fp_sa", 
                              output_directory=OUTPUT_DIRECTORY, seed=SEED, 
                             iteration_list=[50, 100, 150], 
                             max_attempts=5000, temperature_list=[1, 10, 50, 100, 250, 500, 1000, 2500, 5000, 10000]) 

fp_sa_run_stats, fp_sa_run_curves = fp_sa.run()
```

# Neural Network Weights

## Data Preprocessing

```python
# Import Datasets
kiva_df = pd.read_csv('data/kiva_loans.csv')
kiva_df.head()
```

```python
# Add new `days_to_fund` feature
kiva_df['days_to_fund'] = (pd.to_datetime(kiva_df.funded_time) - pd.to_datetime(kiva_df.posted_time)).dt.days
kiva_df = kiva_df[kiva_df.days_to_fund >= 0]
```

```python
# Add new count features `female_borrowers` and `male_borrowers`

def recipient_genders(x):
    lst = x.replace(" ", "").split(',') if not pd.isna(x) else []
    return lst.count('female'), lst.count('male')

gender_counts = kiva_df.borrower_genders.apply(recipient_genders).str
kiva_df.loc[:, 'female_borrowers'] = gender_counts[0]
kiva_df.loc[:, 'male_borrowers'] = gender_counts[1]
```

```python
# Add target feature `is_agricultural_loan` to dataframe
kiva_df['is_agricultural_loan'] = np.where(kiva_df['sector']=='Agriculture', True, False)
```

```python
# Select a subset of features to exclude timeseries data
rel_cols = ['days_to_fund', 'funded_amount', 'loan_amount', 'term_in_months', 'repayment_interval','female_borrowers', 'male_borrowers', 'activity', 'country', 'is_agricultural_loan']
kiva_subset = kiva_df[rel_cols].copy()
kiva_subset.head()
```

```python
# Create feature and target dataframes
# - apply one-hot encoding to categorical columns
# - separate the feature and target data
kiva_feature_cols = kiva_subset.drop(columns = ['is_agricultural_loan']).columns
kiva_features = pd.get_dummies(kiva_subset[kiva_feature_cols], prefix='', prefix_sep='')
kiva_target = kiva_subset[['is_agricultural_loan']]
```

```python
# Split the data into train (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(kiva_features, kiva_target, test_size=0.20, random_state=42)
```

```python
# Normalize feature data
scaler = MinMaxScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# One hot encode target values
one_hot = OneHotEncoder()

y_train_hot = one_hot.fit_transform(y_train).todense()
y_test_hot = one_hot.transform(y_test).todense()
```

### Gradient Descent (benchmarking)

```python
# Initialize neural network object and fit object
gd_nn_model = mlrose.NeuralNetwork(hidden_nodes = [16, 16], activation = 'relu', \
                                 algorithm = 'gradient_descent', max_iters = 50, \
                                 bias = True, is_classifier = True, learning_rate = 0.0001, \
                                 early_stopping = True, clip_max = 5, max_attempts = 100, \
                                 random_state = 3)

gd_nn_model.fit(X_train_scaled, y_train_hot)
```

```python
# Predict labels for train set and assess loss
gd_y_train_pred = gd_nn_model.predict(X_train_scaled)
gd_train_loss = log_loss(y_train_hot, gd_y_train_pred)

# Predict labels for test set and assess loss
gd_y_test_pred = gd_nn_model.predict(X_test_scaled)
gd_test_loss = log_loss(y_test_hot, gd_y_test_pred)
```

```python
gd_report = classification_report(y_test_hot, gd_y_test_pred, zero_division=0, output_dict=True)
gd_report = pd.DataFrame(gd_report)
gd_report['algorithm'] = 'gradient_descent'
```
